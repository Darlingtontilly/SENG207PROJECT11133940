```python
import matplotlib.pyplot as plt

features = df.columns.drop('quality')
for feature in features:
   plt.figure(figsize=(10, 6))
   plt.scatter(df[feature], df['quality'])
   plt.xlabel(feature, fontsize=14)
   plt.ylabel('Quality', fontsize=14)
   plt.title(f'Quality vs {feature}', fontsize=18)
   plt.savefig(f'{feature}_vs_quality.png')
   plt.show()
```

```python
import matplotlib.pyplot as plt

features = df.columns.drop('quality')
for i, feature in enumerate(features):
  plt.figure(figsize=(10, 6))
  plt.scatter(df[feature], df['quality'])
  plt.xlabel(feature, fontsize=14)
  plt.ylabel('Quality', fontsize=14)
  plt.title(f'Quality vs {feature}', fontsize=18)
  plt.savefig(f'{feature}_vs_quality.png')
  plt.show()
```

#### FOR ALCOHOL VS QUALITY
The scatter plot shows the relationship between alcohol content and quality. The x-axis represents the alcohol content, ranging from 8 to 15, and the y-axis represents the quality, ranging from 3 to 8. From the plot, it appears that there is no significant correlation between alcohol content and quality. The data points are scattered across the graph, with most of them concentrated between the quality ratings of 5 to 7. This indicates that most samples have a moderate quality rating, in spite of their alcohol content.


## FOR CHLORIDES VS QUALITY. 
The x-axis represents the level of chlorides, ranging from 0.0 to 0.6, and the y-axis represents the quality, ranging from 3 to 8. From the plot, it appears that there is a negative correlation between the level of chlorides and the quality. Most of the data points are clustered at lower chloride levels, particularly under 0.2. As the level of chlorides increases, the quality tends to decrease. This suggests that wines with higher levels of chlorides tend to have lower quality ratings.


## FOR THE LEVEL OF CITRIC ACID AND QUALITY. 
The x-axis represents the level of citric acid, ranging from 0.0 to 1.0, and the y-axis represents the quality, ranging from 3 to 8. From the plot, it appears that there isn't a clear correlation between the level of citric acid and the quality. The data points are distributed horizontally at each quality level, suggesting that the quality does not significantly change with different levels of citric acid.
